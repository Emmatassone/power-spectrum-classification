```python
import matplotlib.pyplot as plt
import pandas as pd
import re

# Step 1: Initialize lists to store extracted data
accuracies = []
n_jobs = []
n_estimators = []
min_samples_leaf = []
min_samples_split = []

# Step 2: Read the file and extract data
with open('./RF/RF_accuracies_no-split.txt', 'r') as file:
    for line in file:
        if line.startswith("Accuracy"):
            acc = float(re.findall(r"Accuracy on the test set: (\d\.\d+)", line)[0])
            accuracies.append(acc)
        elif "n_jobs" in line:
            n_jobs.append(int(re.findall(r"n_jobs : (\d+)", line)[0]))
        elif "n_estimators" in line:
            n_estimators.append(int(re.findall(r"n_estimators: (\d+)", line)[0]))
        elif "min_samples_leaf" in line:
            min_samples_leaf.append(int(re.findall(r"min_samples_leaf: (\d+)", line)[0]))
        elif "min_samples_split" in line:
            min_samples_split.append(int(re.findall(r"min_samples_split: (\d+)", line)[0]))

# Step 3: Create a DataFrame to organize the data
data = pd.DataFrame({
    'Accuracy': accuracies,
    'n_jobs': n_jobs,
    'n_estimators': n_estimators,
    'min_samples_leaf': min_samples_leaf,
    'min_samples_split': min_samples_split
})

# Step 4: Plot the data using matplotlib directly
plt.figure(figsize=(10,6))
plt.scatter(data['n_estimators'], data['Accuracy'], marker='o', linestyle='-', label='Accuracy')

# Step 5: Annotate each point with its hyperparameter combination (ternion)
for i in range(len(data)):
    label = f"min_leaf={data['min_samples_leaf'][i]}, min_split={data['min_samples_split'][i]}"
    plt.text(data['n_estimators'][i], data['Accuracy'][i], label, fontsize=8, ha='left')

# Customize the plot
plt.title('Accuracy vs n_estimators with Hyperparameters')
plt.xlabel('n_estimators')
plt.ylabel('Accuracy')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

We can see that accuracy is too high independantly of the hyperparameter selection. This suggest the training process is highly biased and is overfitting the dataset
